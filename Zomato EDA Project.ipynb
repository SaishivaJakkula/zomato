```python
df.drop_duplicates(inplace=True)
df
```

```python
df['rate (out of 5)'] = df['rate (out of 5)'].fillna(df['rate (out of 5)'].median())
```

```python
df['avg cost (two people)'] = df['avg cost (two people)'].fillna(
    df['avg cost (two people)'].median()
)
```

```python
df = df.drop(['Unnamed: 0', 'Unnamed: 0.1'], axis=1)
```

```python
df.head()
```

```python
df = df.rename(columns={
    "restaurant name": "restaurant_name",
    "restaurant type": "restaurant_type",
    "rate (out of 5)": "rating",
    "num of ratings": "num_ratings",
    "avg cost (two people)": "avg_cost_two",
    "online_order": "online_order",
    "table booking": "table_booking",
    "cuisines type": "cuisine_type",
    "area": "area",
    "local address": "local_address"
})
```

#### What is the overall distribution of restaurant ratings?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns



plt.figure(figsize=(6,4))
sns.histplot(df["rating"], bins=10, kde=True)
plt.title("Distribution of Restaurant Ratings")
plt.xlabel("Rating (out of 5)")
plt.ylabel("Count")
plt.show()

print(df["rating"].describe())

#Visualization:

# Histogram + KDE

# Shows how ratings are spread (e.g., mostly 3.5–4.5).

# Use describe() to explain mean, median, min, max.
```

#### What is the distribution of average cost for two people?

```python
plt.figure(figsize=(6,4))
sns.histplot(df["avg_cost_two"], bins=15, kde=True)
plt.title("Distribution of Average Cost for Two")
plt.xlabel("Average Cost (two people)")
plt.ylabel("Count")
plt.show()

# Check skew + outliers
df["avg_cost_two"].describe()
```

#### Does higher cost mean better ratings?

```python
plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x="avg_cost_two", y="rating", alpha=0.5)
plt.title("Relation Between Cost for Two and Rating")
plt.xlabel("Average Cost (two people)")
plt.ylabel("Rating")
plt.show()

# Correlation value
corr = df["avg_cost_two"].corr(df["rating"])
print("Correlation between avg cost and rating:", corr)
```

#### Do restaurants with online ordering have better ratings and more reviews?

```python
# Ratings by online_order
plt.figure(figsize=(6,4))
sns.boxplot(data=df, x="online_order", y="rating")
plt.title("Ratings vs Online Order Availability")
plt.xlabel("Online Order Available")
plt.ylabel("Rating")
plt.show()

# Number of ratings by online_order
plt.figure(figsize=(6,4))
sns.boxplot(data=df, x="online_order", y="num_ratings")
plt.title("Number of Ratings vs Online Order Availability")
plt.xlabel("Online Order Available")
plt.ylabel("Number of Ratings")
plt.show()

# Grouped statistics
print(df.groupby("online_order")[["rating", "num_ratings"]].mean())
```

#### Does table booking impact ratings?

```python
plt.figure(figsize=(6,4))
sns.boxplot(data=df, x="table_booking", y="rating")
plt.title("Ratings vs Table Booking Availability")
plt.xlabel("Table Booking Available")
plt.ylabel("Rating")
plt.show()

print(df.groupby("table_booking")["rating"].mean())
```

#### Top 10 cuisine types
